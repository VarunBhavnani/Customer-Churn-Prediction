```python
import pandas as pd
Churn = pd.read_csv('D:/Spring 2022/6337.002 - Predictive/HW5/Churn_telecom.csv')
print('The shape of our features is:', Churn.shape)
Churn.describe()
```

```python
pd.options.display.max_rows = 173
Churn.isnull().sum().sort_values(ascending=False)
```

```python
Churn = Churn.drop(['mailflag','solflag','crtcount','retdays','tot_ret','tot_acpt','REF_QTY','wrkwoman','educ1','rmmou',
                    'rmrev','rmcalls','pcowner','div_type','occu1','proptype','cartype','children','mailordr','mailresp',
                    'pre_hnd_price','last_swap','numbcars','dwllsize','HHstatin','ownrent','dwlltype','lor','income',
                    'adults','infobase','hnd_webcap','prizm_social_one','Customer_ID'],axis=1)
```

```python
Churn.shape
```

```python
Churn.select_dtypes(include=['object']).columns.tolist()
```

```python
#Replace the missing values for categorical columns with mode
Churn['new_cell'] = Churn['new_cell'].fillna(Churn['new_cell'].mode()[0])
Churn['crclscod'] = Churn['crclscod'].fillna(Churn['crclscod'].mode()[0])
Churn['asl_flag'] = Churn['asl_flag'].fillna(Churn['asl_flag'].mode()[0])
Churn['csa'] = Churn['csa'].fillna(Churn['csa'].mode()[0])
Churn['area'] = Churn['area'].fillna(Churn['area'].mode()[0])
Churn['dualband'] = Churn['dualband'].fillna(Churn['dualband'].mode()[0])
Churn['refurb_new'] = Churn['refurb_new'].fillna(Churn['refurb_new'].mode()[0])
Churn['marital'] = Churn['marital'].fillna(Churn['marital'].mode()[0])
Churn['ethnic'] = Churn['ethnic'].fillna(Churn['ethnic'].mode()[0])
Churn['kid0_2'] = Churn['kid0_2'].fillna(Churn['kid0_2'].mode()[0])
Churn['kid3_5'] = Churn['kid3_5'].fillna(Churn['kid3_5'].mode()[0])
Churn['kid6_10'] = Churn['kid6_10'].fillna(Churn['kid6_10'].mode()[0])
Churn['kid11_15'] = Churn['kid11_15'].fillna(Churn['kid11_15'].mode()[0])
Churn['kid16_17'] = Churn['kid16_17'].fillna(Churn['kid16_17'].mode()[0])
Churn['creditcd'] = Churn['creditcd'].fillna(Churn['creditcd'].mode()[0])
Churn['car_buy'] = Churn['car_buy'].fillna(Churn['car_buy'].mode()[0])
```

```python
Num = Churn.select_dtypes(exclude=['object']).columns.tolist()

for x in Num:
   Churn[x] = Churn[x].fillna(Churn[x].median()) 
```

```python
Churn.to_csv('Churn_new.csv')
```
